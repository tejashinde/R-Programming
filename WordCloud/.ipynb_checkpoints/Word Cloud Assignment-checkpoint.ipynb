## Loading Libraries

```r
#install.packages(c('tm','wordcloud','RColorBrewer','corpus'))
library("tm")
library("SnowballC")
library("wordcloud")
library("RColorBrewer")
library("corpus")
```

## Loading the dataset

```r
data = readLines('./../Data/resume.txt')
```

## Making Corpus (Sets of data) and cleaning the data using stopwords, removePunctuation and stripWhitespace

```r
corpusData = Corpus(VectorSource(data))
inspect(corpusData)
corpusData = tm_map(corpusData , content_transformer(tolower))
corpusData = tm_map(corpusData , removeWords, stopwords("english"))
corpusData = tm_map(corpusData , removePunctuation)
corpusData = tm_map(corpusData , stripWhitespace)
corpusData = tm_map(corpusData , removeNumbers)
```

## Mining the data, removing the meaningful words and making a dataframe of top 10 words and their frequencies

```r
minedData = TermDocumentMatrix(corpusData)
datamatrix = as.matrix(minedData)
datavector = sort(rowSums(datamatrix) , decreasing = TRUE)
dataranked = data.frame(word = names(datavector) , freq = datavector)
head(dataranked , 10)
```

## Wordcloud creation

```r
set.seed(1234)
wordcloud(words = dataranked$word, freq = dataranked$freq, min.freq = 1,
          max.words = 200, random.order = FALSE, rot.per = 0.25, 
          colors = brewer.pal(8 , "Dark2"))
```
